```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
THETA_1 = 4
THETA_0 = 1
```

```python
X = np.random.uniform(low=0, high=10, size=(200,))
Y = X * THETA_1 + THETA_0 + np.random.normal(loc=0, scale=2, size=X.shape)
```

```python
GROUND_TRUTH_LINE = X * THETA_1 + THETA_0
```

```python
def plot_data_and_line(theta_1, theta_0):
    preds = X * theta_1 + theta_0
    fig, ax = plt.subplots()
    
    plt.scatter(X, Y, color='blue', s=4, label='data')
    plt.plot(X, GROUND_TRUTH_LINE, color='black', label='ground truth')
    plt.plot(X, preds, color='red', label='predicted line')
    
    plt.legend()
    plt.show()
```

```python
def grad_theta_0(theta_1, theta_0):
    preds = X * theta_1 + theta_0
    grad = np.mean(preds - Y)
    
    return grad

def grad_theta_1(theta_1, theta_0):
    preds = X * theta_1 + theta_0
    error = preds - Y
    grad = error * X
    grad = np.mean(grad)
    
    return grad
```

```python
def gradient_descent_step(theta_1, theta_0, lr):
    val_grad_theta_1 = grad_theta_1(theta_1, theta_0)
    val_grad_theta_0 = grad_theta_0(theta_1, theta_0)
    
    theta_1 = theta_1 - lr * val_grad_theta_1
    theta_0 = theta_0 - lr * val_grad_theta_0
    
    return theta_1, theta_0
```

```python
def train(steps, lr, log=True):
    theta_1, theta_0 = 0, 0
    if log:
        plot_data_and_line(theta_1, theta_0)
    
    for i in range(1, steps):
        theta_1, theta_0 = gradient_descent_step(theta_1, theta_0, lr)
        if log:
            plot_data_and_line(theta_1, theta_0)
        
    return theta_1, theta_0
```

```python
theta_1, theta_0 = train(10, 0.01)
```
